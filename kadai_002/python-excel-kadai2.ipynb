pandasを用いて、以下の要件を満たしたプログラムを記述してください。

まず、元となる社員ごとの売上データは以下の通りです。
日付	社員名	売上	部門

2023-05-17	山田	100	メーカー

2023-05-18	佐藤	200	代理店

2023-05-19	鈴木	150	メーカー

2023-05-20	田中	300	商社

2023-05-21	高橋	250	代理店

「平均売上」列を作成し、全体の売上の平均を算出します。

「業績ランク」列を作成し、以下の条件分岐をして業績ランクを設定します。
関数「performance」を定義して業績ランクの設定をしてください。

業績ランク	条件

A	平均売上+50以上の場合

B	平均売上以上の場合

C	平均売上未満の場合

新規ファイル「業績.xlsx」に書き出します。

```python
from numpy._core.multiarray import result_type
#import pandas
import pandas as pd


# 売上データをCSVファイルから読み込み
df = pd.read_csv('sales.csv', encoding='utf-8')

# Excelファイルを作成する
writer = pd.ExcelWriter('業績.xlsx')

df['平均売上'] = df['売上'].mean()

# スコアを求める関数を定義
def score(rank):
  result_= ''
  if rank >= df['平均売上'].mean() + 50:
    result_ = 'A'
  elif rank >= df['平均売上'].mean():
    result_ = 'B'
  else:
    result_ = 'C'
  return result_

# スコア列を作成し、スコアを求める関数を適用して値を設定
df['業績ランク'] = df['売上'].apply(score)

#エクセルに書き出し
df.to_excel(writer, sheet_name='Sheet1', index=False)

#エクセルファイルを保存
writer.close()


```
