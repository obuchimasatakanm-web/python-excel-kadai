```python
import pandas as pd
import openpyxl
# openpyxl.stylesモジュールのPatternFillクラスをインポート
from openpyxl.styles import PatternFill

# excel read
df1 = pd.read_excel('2022_年間売上表.xlsx')
df2 = pd.read_excel('2023_年間売上表.xlsx')

#連結
# 2つのDataFrameを垂直方向に連結
df_combined = pd.concat([df1, df2], ignore_index=True)

# 売上年・商品ごとに金額を集約
df_aggregated = df_combined.groupby(['売上年', '商品'])['金額（千円）'].sum().reset_index()

df_result = df_aggregated.sort_values(by='商品', ascending=True)

# Excelファイルを作成する
writer = pd.ExcelWriter('売上集計表.xlsx')

# 結合したDataFrameを単一のシートに書き出し
df_result.to_excel(writer, sheet_name='全データ', index=False)

# エクセルファイルを保存
writer.close()

# pandasで作成したExcelファイルをopenpyxlで再度読み込む
workbook = openpyxl.load_workbook('売上集計表.xlsx')

# アクティブなシートを変数「sheet」に格納
sheet = workbook.active

# A1セルの背景色をグレーに設定 (aRGB形式に修正)
sheet['A1'].fill = PatternFill(patternType='solid', fgColor='FFF2F2F2')
sheet['B1'].fill = PatternFill(patternType='solid', fgColor='FFF2F2F2')
sheet['C1'].fill = PatternFill(patternType='solid', fgColor='FFF2F2F2')

# 保存
workbook.save('売上集計表.xlsx')
```
