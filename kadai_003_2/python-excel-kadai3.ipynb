```python
import pandas as pd
import openpyxl
# openpyxl.stylesモジュールのPatternFillクラスをインポート
from openpyxl.styles import PatternFill

# excel read
df1 = pd.read_excel('2022_年間売上表.xlsx')
df2 = pd.read_excel('2023_年間売上表.xlsx')

# 商品ごとに金額を集約
result1 = df1.groupby('商品')['金額（千円）'].sum()
result2 = df2.groupby('商品')['金額（千円）'].sum()

# print(result1)
# print(result2)

# 集約して一度保存
# Excelファイルを作成する
writer1 = pd.ExcelWriter('2022_金額集約後.xlsx')
writer2 = pd.ExcelWriter('2023_金額集約後.xlsx')

# result1 SeriesをDataFrameに変換し、'売上年'、'商品'、'金額（千円）'カラムを設定
df_result1 = result1.reset_index()
df_result1['売上年'] = 2022
df_result1 = df_result1[['売上年','商品', '金額（千円）']]

# df_result1をExcelに書き出し
df_result1.to_excel(writer1, sheet_name='集約データ', index=False, header=True)
# エクセルファイルを保存
writer1.close()

# result2 SeriesをDataFrameに変換し、'売上年'、'商品'、'金額（千円）'カラムを設定
df_result2 = result2.reset_index()
df_result2['売上年'] = 2023
df_result2 = df_result2[[ '売上年','商品', '金額（千円）']]

# df_result2をExcelに書き出し
df_result2.to_excel(writer2, sheet_name='集約データ', index=False, header=True)
# エクセルファイルを保存
writer2.close()

# excel read
df3 = pd.read_excel('2022_金額集約後.xlsx')
df4 = pd.read_excel('2023_金額集約後.xlsx')

#連結
# 2つのDataFrameを垂直方向に連結
df_combined = pd.concat([df3, df4], ignore_index=True)

# 商品ごとに並び替え
df_combined = df_combined.sort_values(by='商品').reset_index(drop=True)

# Excelファイルを作成する
writer = pd.ExcelWriter('売上集計表.xlsx')

# 結合したDataFrameを単一のシートに書き出し
df_combined.to_excel(writer, sheet_name='全データ', index=False)

# エクセルファイルを保存
writer.close()

# pandasで作成したExcelファイルをopenpyxlで再度読み込む
workbook = openpyxl.load_workbook('売上集計表.xlsx')

# アクティブなシートを変数「sheet」に格納
sheet = workbook.active

# A1セルの背景色をグレーに設定 (aRGB形式に修正)
sheet['A1'].fill = PatternFill(patternType='solid', fgColor='FFF2F2F2')
sheet['B1'].fill = PatternFill(patternType='solid', fgColor='FFF2F2F2')

# 保存
workbook.save('売上集計表.xlsx')
```
