```python
# インポート
import openpyxl
import datetime

# エクセルの新規作成
workbook = openpyxl.Workbook()

# アクティブなシートを変数「sheet」に格納
sheet = workbook.active

#　固定文言を指定のセルに追加
sheet['B2'].value = '請求書'
sheet['B4'].value = '株式会社ABC'
sheet['B5'].value = '〒101-0022 東京都千代田区神田練塀町300'
sheet['B6'].value = 'TEL:03-1234-5678 FAX:03-1234-5678'
sheet['B7'].value = '担当者名:鈴木一郎 様'


sheet['F4'].value = 'No.'
sheet['G4'].value = '0001'
sheet['F5'].value = '日付'

now = datetime.datetime.now()
sheet['G5'].value = now.strftime('%Y/%m/%d')

# 数値欄など
sheet['B10'].value = '商品名'
sheet['C10'].value = '数量'
sheet['D10'].value = '単価'
sheet['E10'].value = '金額'

sheet['B11'].value = '商品A'
sheet['C11'].value = 2
sheet['D11'].value = 10000
sheet['E11'].value = sheet['C11'].value * sheet['D11'].value

sheet['B12'].value = '商品B'
sheet['C12'].value = 1
sheet['D12'].value = 15000
sheet['E12'].value = sheet['C12'].value * sheet['D12'].value

# 小計
sheet['B15'].value = '小計'
sheet['E13'].value = sheet['E12'].value + sheet['E11'].value
sheet['E15'].value = sheet['E12'].value + sheet['E11'].value


# 消費税
sheet['B16'].value = '消費税'
sheet['E16'].value = sheet['E15'].value * 0.1

# 合計
sheet['B17'].value = '合計'
sheet['E17'].value = sheet['E15'].value + sheet['E16'].value

# エクセルの名前
name = '請求書_' + now.strftime('%Y%m%d') + '.xlsx'

# エクセルの保存
workbook.save(name)



```
